```
def tratamento(p1,p2):
  arquivo = p1
  temporada = p2

  with open(arquivo) as arq:
    data = json.load(arq)

    GOTA = pd.DataFrame(data['Episodes'])

    GOTA[ 'imdbRating' ] = GOTA['imdbRating'].astype(float)

    GOTB = GOTA.assign(season = temporada).copy(deep=True)

    return GOTB
```
